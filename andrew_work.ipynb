```python
#Imports
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as st
import numpy as np
import warnings
```

With data on all of the MCU movies up through 2022, we intend to see whether or not trends would suggest that the MCU has peaked in popularity and whether Iger's claims have validity.

```python
# Settings the warnings to be ignored
warnings.filterwarnings('ignore')
```

```python
#Read CSV
csv_data = pd.read_csv("./Resources/What is The Best Performing Marvel Movie PUBLIC DATA - Marvel Movies.csv")
```

```python
#Put into a dataframe
MCU_df = pd.DataFrame(csv_data)
MCU_df
```

```python
#Convert % columns to floats
MCU_df_1 = MCU_df
MCU_df_1["critics % score"] = MCU_df_1["critics % score"].str.rstrip('%').astype(float)
MCU_df_1["audience % score"] = MCU_df_1["audience % score"].str.rstrip('%').astype(float)
MCU_df_1.head()
```

```python
#Isolate the desired data
reviews_df = MCU_df[["film", "category", "worldwide gross ($m)", "critics % score", "audience % score", "domestic gross ($m)", \
                     "international gross ($m)", "opening weekend ($m)"]]
reviews_df
```

## Results

By Andrew Prozorovsky

```python
#Profits by franchise
profits_franchise_df = reviews_df.groupby("category").mean()
print("Profits by Marvel Franchise")
display(profits_franchise_df.sort_values("worldwide gross ($m)", ascending=False))
```

The Avengers franchise has the highest average worldwide gross. Their unique category (stand-alone films) have the lowest average worldwide gross. Captain America, despite being the fan-favorite Avenger in poll-data, is low along with Thor and Ant-Man. This data does not include movies from 2023.

```python
#Reviews by franchise
reviews_franchise_df = reviews_df.groupby("category").mean()
print("Reviews by Marvel Franchise")
display(reviews_franchise_df.sort_values("critics % score", ascending=False))
```

Spider-Man has the highest average critic score and their Unique category (Stand-alone films) again has the lowest in both critics and audience score averages. The Avengers franchise falls to 4th in this category.

```python
#Graph critics % score against audience % score
groups = reviews_df.groupby("category")
for name, group in groups:
    plt.plot(group["critics % score"], group["audience % score"], marker="o", linestyle="", label=name, markeredgecolor='black')
plt.legend(fontsize='small')
plt.xlabel("MCU Critics % Scores")
plt.ylabel("MCU Audience % Scores")
plt.title("Audience % Scores vs. Critics % Scores for MCU Movies")
(slope, intercept, rvalue, pvalue, stderr) = st.linregress(reviews_df["critics % score"], reviews_df["audience % score"])
correlation = st.pearsonr(reviews_df["critics % score"], reviews_df["audience % score"])
regress_values = reviews_df["critics % score"] * slope + intercept
line_eq = f"y = {str(round(slope, 2))}x + {str(round(intercept, 2))}"
plt.plot(reviews_df["critics % score"], regress_values, color="red")
plt.annotate(line_eq, (50,90), fontsize=15, color="red")
print(f"The r-value between audience reviews and critics reviews is {round(correlation[0], 2)}.")
plt.savefig("Outputs/Fig1.png")
```

Audience scores versus critics scores are highly correlated but still tend to differ. The outlier on the bottom was Captain Marvel and the outlier on the left was Eternals.

```python
#Graph opening weekend by year
groups = MCU_df.groupby("category")
for name, group in groups:
    plt.plot(group["year"], group["opening weekend ($m)"], marker="o", linestyle="", label=name, markeredgecolor='black')
plt.legend(fontsize='small')
plt.xlabel("Year")
plt.ylabel("Opening Weekend Box Office")
plt.title("Opening Weekend Box Office by Year for MCU Movies")
(slope, intercept, rvalue, pvalue, stderr) = st.linregress(MCU_df["year"], MCU_df["opening weekend ($m)"])
correlation = st.pearsonr(MCU_df["year"], MCU_df["opening weekend ($m)"])
regress_values = MCU_df["year"] * slope + intercept
line_eq = f"y = {str(round(slope, 2))}x + {str(round(intercept, 2))}"
plt.plot(MCU_df["year"], regress_values, color="red")
plt.annotate(line_eq, (2015,300), fontsize=15, color="red")
print(f"The r-value between year and opening weekend box office is {round(correlation[0], 2)}.")
plt.savefig("Outputs/Fig2.png")
```

Opening Weekend's have gotten more profitable for Marvel as it built the MCU brand. But post Avengers: Endgame may suggest that a linear regression is not the best fit here.

```python
#Graph Worldwide Gross against Critics % Score
groups = reviews_df.groupby("category")
for name, group in groups:
    plt.plot(group["critics % score"], group["worldwide gross ($m)"], marker="o", linestyle="", label=name, markeredgecolor='black')
plt.legend(fontsize='small')
plt.title("Worldwide Gross ($Millions) vs. Critics % Scores for MCU Movies")
plt.ylabel("Worldwide Gross ($Millions)")
plt.xlabel("MCU Critics % Scores")
(slope, intercept, rvalue, pvalue, stderr) = st.linregress(reviews_df["critics % score"], reviews_df["worldwide gross ($m)"])
correlation = st.pearsonr(reviews_df["critics % score"], reviews_df["worldwide gross ($m)"])
regress_values = reviews_df["critics % score"] * slope + intercept
line_eq = f"y = {str(round(slope, 2))}x + {str(round(intercept, 2))}"
plt.plot(reviews_df["critics % score"], regress_values, color="red")
plt.annotate(line_eq, (50,1200), fontsize=15, color="red")
print(f"The r-value between worldwide gross (millions) and critics reviews is {round(correlation[0], 2)}.")
plt.savefig("Outputs/Fig3.png")
```

Critics scores have had effects on the worldwide revenue for Marvel movies. Or maybe worldwide revenues pressured critics into giving better ratings? Perhaps both of these are affected by a counfounding variable. The relationship may not be so obvious.

```python
#Graph Worldwide Gross against Audience % Score
groups = reviews_df.groupby("category")
for name, group in groups:
    plt.plot(group["audience % score"], group["worldwide gross ($m)"], marker="o", linestyle="", label=name, markeredgecolor='black')
plt.legend(fontsize='small')
plt.title("Worldwide Gross ($millions) vs. Audience % Scores for MCU Movies")
plt.ylabel("Worldwide Gross ($Millions)")
plt.xlabel("MCU Audience % Scores")
(slope, intercept, rvalue, pvalue, stderr) = st.linregress(reviews_df["audience % score"], reviews_df["worldwide gross ($m)"])
correlation = st.pearsonr(reviews_df["audience % score"], reviews_df["worldwide gross ($m)"])
regress_values = reviews_df["audience % score"] * slope + intercept
line_eq = f"y = {str(round(slope, 2))}x + {str(round(intercept, 2))}"
plt.plot(reviews_df["audience % score"], regress_values, color="red")
plt.annotate(line_eq, (60,2500), fontsize=15, color="red")
print(f"The r-value between worldwide gross (millions) and audience reviews is {round(correlation[0], 2)}.")
plt.savefig("Outputs/Fig4.png")
```

The same logic applies here. At first glance, the relationship seems to suggest that audience reviews and spreading the word really does excite or dampen a movie's profits. But a potential alternative would be that as more people see the movie and rate it, the scores tend to steady-state at a higher review score.

```python
#Graph of Critics Review Scores vs. Opening Weekend
groups = reviews_df.groupby("category")
for name, group in groups:
    plt.plot(group["critics % score"], group["opening weekend ($m)"], marker="o", linestyle="", label=name, markeredgecolor='black')
plt.legend(fontsize='small')
plt.title("Opening Weekend ($Millions) vs. Critics % Scores for MCU Movies")
plt.ylabel("Opening Weekend ($Millions)")
plt.xlabel("Critics % Scores")
(slope, intercept, rvalue, pvalue, stderr) = st.linregress(reviews_df["critics % score"], reviews_df["opening weekend ($m)"])
correlation = st.pearsonr(reviews_df["critics % score"], reviews_df["opening weekend ($m)"])
regress_values = reviews_df["critics % score"] * slope + intercept
line_eq = f"y = {str(round(slope, 2))}x + {str(round(intercept, 2))}"
plt.plot(reviews_df["critics % score"], regress_values, color="red")
plt.annotate(line_eq, (70, 300), fontsize=15, color="red")
print(f"The r-value between opening weekend box office and critics reviews is {round(correlation[0], 2)}.")
plt.savefig("Outputs/Fig5")
```

As review embargos lift, is opening weekend a game of momentum and turnout? Positive critics scores appear to have increased box office traffic.

```python
#Graph of Audience Review Scores vs. Opening Weekend
groups = reviews_df.groupby("category")
for name, group in groups:
    plt.plot(group["audience % score"], group["opening weekend ($m)"], marker="o", linestyle="", label=name, markeredgecolor='black')
plt.legend(fontsize='small')
plt.title("Opening Weekend ($Millions) vs. Audience % Scores for MCU Movies")
plt.ylabel("Opening Weekend ($Millions)")
plt.xlabel("Audience % Scores")
(slope, intercept, rvalue, pvalue, stderr) = st.linregress(reviews_df["audience % score"], reviews_df["opening weekend ($m)"])
correlation = st.pearsonr(reviews_df["audience % score"], reviews_df["opening weekend ($m)"])
regress_values = reviews_df["audience % score"] * slope + intercept
line_eq = f"y = {str(round(slope, 2))}x + {str(round(intercept, 2))}"
plt.plot(reviews_df["audience % score"], regress_values, color="red")
plt.annotate(line_eq, (60,300), fontsize=15, color="red")
print(f"The r-value between opening weekend box office and audience reviews is {round(correlation[0], 2)}.")
plt.savefig("Outputs/Fig6")
```

Higher opening weekend revenues and audience ratings also correlated with one another positively.
