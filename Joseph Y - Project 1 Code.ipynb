```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load data from csv
movie_data = pd.read_csv('~/Downloads/What is The Best Performing Marvel Movie PUBLIC DATA - Marvel Movies (1).csv')
```

```python
#   Calculate statistics
data_comparison = movie_data[['domestic gross ($m)', 'international gross ($m)']]

domestic_stats = data_comparison['domestic gross ($m)'].describe()
international_stats = data_comparison['international gross ($m)'].describe()

grouped_data = movie_data.groupby('category')
grouped_statistics = grouped_data.agg({
    'worldwide gross ($m)': 'sum',
    'domestic gross ($m)': 'sum',
    'international gross ($m)': 'sum'
})
print(grouped_statistics)
```

```python
# Data Visualization
plt.figure(figsize=(10, 6))
plt.bar(['Domestic', 'International'], [domestic_stats['mean'], international_stats['mean']])
plt.title('Average Earnings for Domestic vs International Movies')
plt.ylabel('Average Earnings')
plt.show()
```

```python
# Data Visualization - Scatter Chart
# Create a scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(grouped_statistics['domestic gross ($m)'], grouped_statistics['international gross ($m)'], c='red', label='Scatter Plot')
plt.title('Scatter Plot of Domestic vs International Earnings')
plt.xlabel('Domestic Earnings')
plt.ylabel('International Earnings')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#   Data Visualization - Pie Chart
# Calculate total domestic gross earnings for each category
grouped_domestic_earnings = grouped_data['domestic gross ($m)'].sum()

# Create a pie chart
labels = grouped_domestic_earnings.index
sizes = grouped_domestic_earnings.values
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'orange']

plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=140)
plt.title('Distribution of Domestic Gross Earnings by Category')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
# Calculate total international gross earnings for each category
grouped_international_earnings = grouped_data['international gross ($m)'].sum()

# Create a pie chart
labels = grouped_international_earnings.index
sizes = grouped_international_earnings.values
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'orange']

plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=140)
plt.title('Distribution of International Gross Earnings by Category')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
# Create histograms
plt.figure(figsize=(10, 6))
for category, group in movie_data.groupby('category'):
    plt.hist([group['domestic gross ($m)'], group['international gross ($m)']],
             bins=20, alpha=0.5, label=['Domestic', 'International'])
plt.title('Histogram of Gross Earnings by Category')
plt.ylabel('Frequency')
plt.xlabel('Gross Earnings')
plt.legend()
plt.show()
```

```python
# Create a scatter plot by release year
plt.figure(figsize=(10, 6))
#plt.scatter(movie_data['year'], movie_data['domestic gross ($m)'], label='Domestic', alpha=0.7)
plt.scatter(movie_data['year'], movie_data['international gross ($m)'], label='International', alpha=0.7)
plt.title('Scatter Plot of International Gross Earnings by Release Year')
plt.xlabel('Release Year')
plt.ylabel('Gross Earnings')
plt.legend()
plt.grid(True)
plt.show()
```
